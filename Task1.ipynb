```python
df['Aggregate rating']=pd.to_numeric(df['Aggregate rating'],errors='coerce')
```

```python
mean=df['Aggregate rating'].mean()
median=df['Aggregate rating'].median()
mode=df['Aggregate rating'].mode()

print('Mean Rating',mean)
print('Median Rating',median)
print('Mode Rating',mode)
```

```python
plt.hist(df['Aggregate rating'],bins=20,color='orange')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Aggregate Rating')
plt.show()
```
