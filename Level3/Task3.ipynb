```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv(r'c:\Users\Dell\Downloads\Dataset .csv')
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['Aggregate rating'],bins=20,color='skyblue')
plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=df['Cuisines'], y=df['Aggregate rating'], data=df, ci=None)
plt.xticks(rotation=45)
plt.title('Average Rating by Cuisine Type')
plt.xlabel('Cuisine Type')
plt.ylabel('Average Rating')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df['City'], y=df['Aggregate rating'], data=df, ci=None)
plt.xticks(rotation=45)
plt.title('Average Rating by City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.show()
```
