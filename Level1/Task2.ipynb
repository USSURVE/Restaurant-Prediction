```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'c:\Users\Dell\Downloads\Dataset .csv')
```

```python
df.head()
```

```python
x=df['Latitude']
```

```python
mean=x.mean()
mode=x.mode()
mediian=x.median()
std=x.std()
print('Latitude mean is',mean)
print('Latitude mode is',mode)
print('Latitude median is',mediian)
print('Std of latitude is ',std)
```

```python
y=df['Longitude']
mean=y.mean()
mode=y.mode()
median=y.median()
std=y.std()
print('Mean of Longitude is ',mean)
print('Mode of Longitude is ',mode)
print('Median of Longitude is', median)
print('Std of Longitude is ',std)
```

```python
c=df['Price range']
mean=c.mean()
mode=c.mode()
median=c.median()
std=c.std()
print('Mean of Price Range is ',mean)
print('Mode of Price Range is ',mode)
print('Median of Price Range is', median)
print('Std of Price Range is ',std)
```

```python
b=df['Aggregate rating']
mean=b.mean()
mode=b.mode()
median=b.median()
std=b.std()
print('Mean of Aggregate rating is ',mean)
print('Mode of Aggregate rating is ',mode)
print('Median of Aggregate rating is', median)
print('Std of Aggregate rating is ',std)
```

```python
d=df['Votes']
mean=d.mean()
mode=d.mode()
median=d.median()
std=d.std()
print('Mean of  Votes is ',mean)
print('Mode of Votes is ',mode)
print('Median of Votes is', median)
print('Std of Votes is ',std)
```

```python
country_distribution=df['Country Code'].value_counts()
print("Country Distribution :",country_distribution)
```

```python
city_distribution=df['City'].value_counts()
print('City Distribution :',city_distribution)
```

```python
cusine_distribution=df['Cuisines'].value_counts()
print('Cusine Distribution :',cusine_distribution)
```

```python
#for exploring cusines
cusine_counts={}
for cusine in df['Cuisines']:
    cusine_counts[cusine]=cusine_counts.get(cusine,0)+1
```

```python
cusine_counts
```

```python
cusine_distribution=pd.Series(cusine_counts)
```

```python
cusine_distribution=cusine_distribution.sort_values(ascending=False)
```

```python
cusine_distribution
```

```python
cusine_distribution.head(10)
print('Top Cusines are:',cusine_distribution)
```

```python
top_cities=city_distribution.head(10)
print('Top Cites with Highest number of Restaurents are:',top_cities)
```

```python
top_cities.plot(kind='bar',figsize=(10,6),color='blue')
plt.title('Top Cities with Highest Number of Restaurent ')
plt.xlabel('City')
plt.ylabel('Number of Restaurent')
plt.xticks(rotation=45)
plt.show()
```
